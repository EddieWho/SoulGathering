```python
import numpy as np
import matplotlib.pyplot as plt

def insurance_simulation(num_policies, num_simulations, disaster_probability, premium, max_payout):
    bankruptcies = 0

    for _ in range(num_simulations):
        # 计算每个受保人的最大赔付金额
        individual_max_payouts = np.random.uniform(0, max_payout, num_policies)
        
        # 计算实际需要赔付的金额（考虑多个受保人）
        actual_payout = np.sum(np.minimum(individual_max_payouts, max_payout))
        
        # 判断是否破产
        if actual_payout > premium:
            bankruptcies += 1

    bankruptcy_probability = bankruptcies / num_simulations
    return bankruptcy_probability

def main():
    num_simulations = 1000
    disaster_probability = 0.05  # 灾难发生概率
    premium = 80000  # 保金
    max_payout = 300000  # 最大赔付金额

    policy_range = range(100, 5001, 10)  # 保单量范围从100到5000，步长为100
    bankruptcy_probabilities = []

    for num_policies in policy_range:
        # 多次运行模拟，获取平均结果
        probabilities = [insurance_simulation(num_policies, num_simulations, disaster_probability, premium, max_payout) for _ in range(10)]
        average_probability = np.mean(probabilities)
        bankruptcy_probabilities.append(average_probability)

    # 可视化结果
    plt.plot(policy_range, bankruptcy_probabilities)
    plt.title('Insurance Company Bankruptcy Probability vs. Number of Policies')
    plt.xlabel('Number of Policies')
    plt.ylabel('Average Bankruptcy Probability')
    plt.show()

if __name__ == "__main__":
    main()
```
